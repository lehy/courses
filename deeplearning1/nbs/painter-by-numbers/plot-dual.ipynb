```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import sys
sys.path.append('..')
import ronan
import matplotlib.pyplot as plt
```

```python
def plot_models(models, components=['accuracy']):
    fig, axes = plt.subplots(2, figsize=(14, 12))

    for ax, x in zip(axes, ['iteration', 'elapsed_time']):
        ax = [ax]
        for model in models:
            ronan.plot_logs([model], axes=ax, x=x, components=components)
    axes[0].legend()
    return axes

models = [#'ResNet50-res5-1M-batch128',
          'DualNet-1M-batch128',      
          'DualNet-1M-ores2-batch128',
          'DualNet-1M-ores5-fres1-batch128'
]
axes = plot_models(models)
axes[0].set_xlim((0, 100))
```
