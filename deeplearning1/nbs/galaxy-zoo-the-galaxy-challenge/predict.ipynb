```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('..')
import ronan
import chainer
import numpy as np
import pandas as pd
```

```python
datasets = ronan.load_galaxy_datasets()
size_prediction = len(datasets.train[0][1])
net = ronan.ResNet50(output_size=size_prediction)
chainer.serializers.load_npz('res5-batch128/snapshot_model_1050', net)
```

```python
small_dataset = ronan.AttrDict(datasets)
small_dataset.test = chainer.datasets.SubDataset(datasets.test, 0, 512)
small_dataset.test.files = datasets.test.files
prediction = ronan.predict(net, small_dataset)
prediction
```
