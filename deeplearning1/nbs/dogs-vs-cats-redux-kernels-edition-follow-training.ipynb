```python
def read_logs(*result_dirs):
    logs = []
    for result_dir in result_dirs:
        with file('{}/log'.format(result_dir)) as logf:
            logs += json.load(logf)
    log = pd.DataFrame(logs)
    return log

def plot_logs(*result_dirs):
    log = read_logs(*result_dirs)
    for component in ['accuracy', 'loss']:
        plt.figure(figsize=(16, 8))
        plt.plot(log.iteration,
                 log['main/{}'.format(component)],
                 label='train/{}'.format(component))
        plt.plot(log.iteration,
                 log['validation/main/{}'.format(component)].interpolate(),
                 label='validation/{}'.format(component))
        plt.legend()
 
```
